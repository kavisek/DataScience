# <center> Importing Data from RDS
### Import Prelimnaries

```python
# import pymysql and pandas information
import pandas as pd
import pymysql

# setting up db  connection parameters 
host="aws.endpoint.us-east-2.rds.amazonaws.com"
port=3306
db ='world'
user="Username"
password="Password"

# Set up the connection
conn = pymysql.connect(db=db, host=host, user=user,port=port, 
                       passwd=password)
```

### Import and Query the Data

```python
# read sql data with pandas
df = pd.read_sql('SELECT * from country;',con=conn)

# view the dataframe
df.sample(5)
```
